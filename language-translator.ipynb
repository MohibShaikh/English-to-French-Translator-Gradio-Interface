```python
import gradio as gr                  # UI Library
from transformers import pipeline    # Transformers Pipeline
```

```python
translation_pipeline = pipeline("translation", model="Helsinki-NLP/opus-mt-en-fr")
translation_pipeline('hi, thank you')
```

```python
def translator_en_fr(text):
    result = translation_pipeline(text)
    return result[0]['translation_text']
```

```python
translator_en_fr('My name\'s Mohib')
```

```python
interface = gr.Interface(fn=translator_en_fr,
                          inputs=gr.components.Textbox(lines=3, placeholder='Enter Text to translate (EN-FR)'),
                          outputs='text')
```

```python
interface.launch()
```
